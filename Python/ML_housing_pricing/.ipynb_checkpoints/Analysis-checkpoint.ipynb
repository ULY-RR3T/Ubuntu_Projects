```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
%matplotlib inline
```

```python
housing = pd.read_csv('Resources/housing.csv')
housing.info()
```

```python
housing.describe()
```

```python
housing.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
def split_train_test(data,ratio):
    np.random.seed(42)
    shuffled_indicies = np.random.permutation(len(data))
    target_index = int(len(data)*ratio)
    train_indicies = shuffled_indicies[target_index:]
    test_indicies = shuffled_indicies[:target_index]
    return data.iloc[train_indicies],data.iloc[test_indicies]

train_set,test_set = split_train_test(housing,0.2)
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(housing, test_size=0.2, random_state=42)
```

```python
housing["income_cat"] = pd.cut(housing["median_income"],bins = [0.,1.5,3.0,4.5,6,np.inf],labels = [1,2,3,4,5])
housing["income_cat"].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index,test_index in split.split(housing,housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
    
# Sanity check for StratifiedShuffleSplit
print("Strat:\n",(strat_test_set["income_cat"].value_counts()/len(strat_test_set)).sort_index())
print("Normal:\n",(test_set["income_cat"].value_counts()/len(test_set)).sort_index())
(housing["income_cat"].value_counts()/len(housing)).sort_index()
```

```python
for set_ in (strat_train_set,strat_test_set):
    set_.drop("income_cat",axis = 1, inplace = True)

cpy = strat_train_set.copy()
```

```python
cpy.plot(kind = "scatter", x = "longitude", y = "latitude",alpha = 0.1)
```

```python
housing.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
s=housing["population"]/100, label="population", figsize=(10,7),
c="median_house_value", cmap=plt.get_cmap("jet"), colorbar=True,
)
plt.legend()
```
